```python
import pandas as pd
from transformers import XLMRobertaTokenizer

# Load tokenizer
tokenizer = XLMRobertaTokenizer.from_pretrained("/mount/studenten-temp1/users/randomuser/newsreclib_new/newsreclib/local_models/xlm-roberta-base", local_files_only=True)

# Function to trim text to 96 tokens
def trim_to_96_tokens(text):
    tokens = tokenizer.encode(text, truncation=True, max_length=96)
    return tokenizer.decode(tokens, skip_special_tokens=True)
df = pd.read_parquet("./data/npr/articles_small.parquet")
df["combined_text"] = df["title"] + ". " + df["body"]
df["category_str"] = df["topics"].apply(
    lambda x: x[0] if isinstance(x, np.ndarray) and x.size > 0 else "na"
)
# Apply to a column (e.g., 'title' or 'text')
df["combined_text_96"] = df["combined_text"].apply(trim_to_96_tokens)
```
